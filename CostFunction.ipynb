```python
import numpy as np
import matplotlib.pyplot as plt
```

**Implement MSE**

```python
def compute_cost(x, y, w, b):
    n = len(x)
    total_error = 0
    for i in range(n):
        y_pred = m * x[i] + b  
        error = (y_pred - y[i]) ** 2
        total_error += error           
    cost = total_error / (2 * n)
    return cost
```

**Implement Gradient Descent**

```python
def gradient_descent(x, y, m, c, learning_rate, iterations):
    n = len(x)
    loss = []

    for i in range(iterations):
        dm = 0
        dc = 0

        for j in range(n):
            y_pred = m * x[j] + c
            error = y_pred - y[j]
            dm += error * x[j]
            dc += error

        dm = dm / n
        dc = dc / n

        # Update m and c (slope and intercept)
        m = m - learning_rate * dm
        c = c - learning_rate * dc

        # Compute and store cost using updated m and c
        cost = compute_cost(x, y, m, c)
        loss.append(cost)

    return m, c, loss
```

**Initialize Feature, Target, Slope, Intercept, Learning_rate, Iterations**

```python
x = list(map(float, input("Enter values of x (space-separated): ").split())) # x = [1, 2, 3, 4, 5]
y = list(map(float, input("Enter values of y (space-separated): ").split())) # y = [3, 5, 7, 9, 11]

if len(x) != len(y):
    raise ValueError("x and y must have the same length.")
```

```python
m = float(input("Enter initial value of m (slope): "))  # 0, 2.0
c = float(input("Enter initial value of c (intercept): "))  # 0, 1.0
```

```python
learning_rate = float(input("Enter learning rate (e.g., 0.01): "))
iterations = int(input("Enter number of iterations (e.g., 1000): "))
```

```python
final_m, final_c, mse = gradient_descent(x, y, m, c, learning_rate, iterations)

print("Final slope :", final_m)
print("Final intercept :", final_c)
print("Final cost:", mse[-1])
```

**Plot cost function**

```python
plt.plot(mse)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost vs Iterations")
plt.grid(True)
plt.show()
```

**Best fit line**

```python
# Plot training data
plt.scatter(x, y, color='red', label='Training Data')

# Plot regression line
y_pred_line = [m * xi + c for xi in x]
plt.plot(x, y_pred_line, label='Prediction Line')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid(True)
plt.show()
```
