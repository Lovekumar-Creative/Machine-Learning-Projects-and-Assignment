```python
x = dataframe.drop(columns='Modal Price', axis=1)
y = dataframe['Modal Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_norm = scaler.fit_transform(x_train)
```

```python
x_train_norm
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.boxplot(data=x_train_norm, ax=ax)
plt.savefig("Agri_boxTrainNorm.png")
```

```python
x_test_norm = scaler.transform(x_test)
```

```python
x_test_norm
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.boxplot(data=x_test_norm, ax=ax)
plt.savefig("agr_boxTestNorm.png")
```

```python
from sklearn.linear_model import LinearRegression
Regression = LinearRegression()
Regression.fit(x_train_norm, y_train)
```

```python
print(Regression.coef_)
```

```python
print(Regression.intercept_)
```

```python
prediction = Regression.predict(x_test_norm)
prediction
```

```python
residual = y_test - prediction
```

```python
residual
```

```python
sns.displot(residual, kind="kde")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print(mean_squared_error(y_test, prediction))
print(mean_absolute_error(y_test, prediction))
print(r2_score(y_test, prediction))
```
